```python
# Import library AI
import numpy as np

# Buat data sederhana (seperti data serangan cyber)
data_serangan = np.array([10, 20, 30, 40, 50])

print("Data serangan:", data_serangan)
print("Rata-rata:", np.mean(data_serangan))
print("Maksimum:", np.max(data_serangan))
print("Minimum:", np.min(data_serangan))
```

```python
# Simulasi deteksi anomali sederhana
# Ini konsep dasar AI Security!
data = np.array([10, 12, 11, 13, 10, 95, 11, 12])
# 95 adalah anomali (serangan!)

rata_rata = np.mean(data)
batas = rata_rata * 2

print("Rata-rata traffic normal:", rata_rata)
print("Batas mencurigakan:", batas)

for nilai in data:
    if nilai > batas:
        print(f"⚠️ ALERT! Traffic mencurigakan terdeteksi: {nilai}")
    else:
        print(f"✅ Normal: {nilai}")
```

```python
# Import library Machine Learning
from sklearn.ensemble import IsolationForest
import numpy as np

# Data traffic jaringan (normal + serangan)
traffic = np.array([
    [10], [12], [11], [13], [10],
    [11], [12], [10], [13], [11],
    [95], [98], [120]  # ini serangan!
])

# Buat model AI untuk deteksi anomali
model = IsolationForest(contamination=0.2)
model.fit(traffic)

# Prediksi mana yang normal, mana yang serangan
hasil = model.predict(traffic)

for i, (nilai, prediksi) in enumerate(zip(traffic, hasil)):
    if prediksi == -1:
        print(f"⚠️ SERANGAN terdeteksi! Traffic: {nilai[0]}")
    else:
        print(f"✅ Normal: {nilai[0]}")
        
```

```python
# Import Pandas - library wajib untuk AI Security!
import pandas as pd
import numpy as np

# Buat data log serangan cyber
data = {
    'waktu': ['08:00', '08:05', '08:10', '08:15', '08:20', '08:25'],
    'ip_address': ['192.168.1.1', '192.168.1.2', '10.0.0.1', '192.168.1.1', '172.16.0.1', '10.0.0.1'],
    'jenis_serangan': ['Normal', 'Normal', 'SQL Injection', 'Brute Force', 'Normal', 'XSS'],
    'traffic': [120, 115, 890, 750, 130, 420]
}

# Buat DataFrame (seperti tabel Excel tapi di Python!)
df = pd.DataFrame(data)
print(df)
```

```python
# Analisis data serangan
print("=== STATISTIK LOG SERANGAN ===")
print("Total kejadian:", len(df))
print("Rata-rata traffic:", df['traffic'].mean())
print("Traffic tertinggi:", df['traffic'].max())
print()

# Hitung jumlah per jenis serangan
print("=== JENIS SERANGAN ===")
print(df['jenis_serangan'].value_counts())
print()

# Filter hanya yang berbahaya
print("=== SERANGAN BERBAHAYA ===")
berbahaya = df[df['jenis_serangan'] != 'Normal']
print(berbahaya)
print()

# IP mana yang paling sering menyerang
print("=== IP PALING MENCURIGAKAN ===")
print(df['ip_address'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Grafik 1 - Traffic per waktu
plt.figure(figsize=(10, 4))
plt.plot(df['waktu'], df['traffic'], marker='o', color='green')
plt.axhline(y=500, color='red', linestyle='--', label='Batas Bahaya')
plt.title('Network Traffic Monitor')
plt.xlabel('Waktu')
plt.ylabel('Traffic')
plt.legend()
plt.grid(True)
plt.show()

# Grafik 2 - Pie chart jenis serangan
plt.figure(figsize=(6, 6))
df['jenis_serangan'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribusi Jenis Serangan')
plt.show()
```
